```python
import pandas as pd
import xgboost
```

```python
dataset = pd.read_csv('final_data.csv')
X = dataset.drop(['output', 'DISTRICT', 'Unnamed: 0'], axis=1)
y = dataset['output']
X
```

```python
import xgboost
xgb = xgboost.XGBRegressor(
    max_depth=2,
    gamma=2,
    eta=0.8,
    reg_alpha=0.5,
    reg_lambda=0.5
)
res = xgb.fit(X,y)
pred_xgb = res.predict(X)
pred_xgb
```

```python
from sklearn.linear_model import LogisticRegression
m = LogisticRegression(
    penalty='l2',
    C=0.1
)
m.fit(X, y)
pred_logregr = m.predict_proba(X)[:,1]
pred_logregr
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X,y)
pred_random = rf.predict_proba(X)[:,1]
pred_random
```

```python
pred = (pred_xgb + pred_logregr + pred_random)/3
pred
```
