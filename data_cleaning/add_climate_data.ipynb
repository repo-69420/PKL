```python
n_districts = ["Achham","Arghakhanchi","Baglung","Baitadi","Bajhang","Bajura","Banke","Bara","Bardiya","Bhaktapur","Bhojpur","Chitwan","Dadeldhura","Dailekh","Dang","Darchula","Dhading","Dhankuta","Dhanusa","Dolakha","Dolpa","Doti","Gorkha","Gulmi","Humla","Ilam","Jajarkot","Jhapa","Jumla","Kailali","Kalikot","Kanchanpur","Kapilvastu","Kaski","Kathmandu","Kavrepalanchok","Khotang","Lalitpur","Lamjung","Mahottari","Makawanpur","Manang","Morang","Mugu","Mustang","Myagdi","Nawalparasi","Nuwakot","Okhaldhunga","Palpa","Panchthar","Parbat","Parsa","Pyuthan","Ramechhap","Rasuwa","Rautahat","Rolpa","Rukum","Rupandehi","Salyan","Sankhuwasabha","Saptari","Sarlahi","Sindhuli","Sindhupalchwok","Sindhupalchok", "Siraha","Solukhumbu","Sunsari","Surkhet","Syangja","Tanahu","Taplejung","Terhathum","Udayapur"]
```

```python
districts = nd['district']
```

```python
nd
```

```python
not_converted = []
for i in range(len(nd)):
    if df['district'][i] not in n_districts:
        not_converted.append(df['district'][i])
```

```python
not_converted
```

```python
data = nd[~nd.district.isin(not_converted)]
```

```python
data
```

```python
data.rename(columns = {"event_year": "YEAR", "event_month": "MONTH", "district":"DISTRICT"}, inplace=True)
```

```python
climate = pd.read_csv('climate_data_nepal_district_wise_monthly.csv')
```

```python
cd = climate.copy()
```

```python
cd
```

```python
cd.drop(['DATE', 'LAT', 'LON'], axis=1, inplace=True)
```

```python
merge = pd.merge(data, cd, how ='inner', on =['YEAR', 'MONTH', 'DISTRICT'])
```
